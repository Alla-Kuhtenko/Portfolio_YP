   - проверить гипотезу различия дохода между теми клиентами, которые ушли и теми, которые остались

## Часть 4. Статистический анализ данных. Проверка гипотез

```python
# уровень статистической значимости
alpha = .05
```

### Гипотеза №1. Отток - доход

**Средний уровень дохода влияет на отток клиентов.**

Н0: Средний уровень дохода клиентов, которые ушли, равен среднему уровню дохода клиентов, которые остались

Н1: Средний уровень дохода клиентов, которые ушли, НЕ равен среднему уровню дохода клиентов, которые остались

```python
# проверка на равенство дисперсий (тест Левене для определения того, имеют ли две или более группы одинаковые отклонения)
stat1, p1 = st.levene(
    good_data.query('churn == 0')['est_salary'], 
    good_data.query('churn == 1')['est_salary']
)
print(f"p-значение:{p1}")
if p1 < alpha:
    print("Дисперсии не равны")
else:
    print("Дисперсии равны")
```

```python
hyp_1 = st.ttest_ind(
    good_data.query('churn == 0')['est_salary'], 
    good_data.query('churn == 1')['est_salary'], 
    equal_var = False     # т.к.дисперсии выборок не равны, это указывается в параметре equal_var
)

print('p-значение: ', hyp_1.pvalue)

if hyp_1.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

### Вывод  
Предположение, что зарплата и остальные источники дохода как-то связаны с уходом клиентов, не подтвердилось.   
Скорее всего, зарплатный продукт не имеет каких-то значимых недостатков в сравнении с другими банками.

###  Гипотеза №2. Отток - наличие кредитной карты

**Наличие кредитных карт дохода влияет на отток клиентов.**

Н0: Средний показатель наличия карт у клиентов, которые ушли, равен среднему показателю наличия кредитных карт клиентов, которые остались

Н1: Средний показатель наличия карт у клиентов, которые ушли, НЕ равен среднему показателю наличия кредитных карт клиентов, которые остались

```python
# проверка на равенство дисперсий (тест Левене для определения того, имеют ли две или более группы одинаковые отклонения)
stat2, p2 = st.levene(
    good_data.query('churn == 0')['credit_card'], 
    good_data.query('churn == 1')['credit_card']
)
print(f"p-значение:{p2}")
if p2 < alpha:
    print("Дисперсии не равны")
else:
    print("Дисперсии равны")
```

```python
hyp_2 = st.ttest_ind(
    good_data.query('churn == 0')['credit_card'], 
    good_data.query('churn == 1')['credit_card'], 
    equal_var = False     # т.к.дисперсии выборок не равны, это указывается в параметре equal_var
)

print('p-значение: ', hyp_1.pvalue)

if hyp_2.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```
